```python
# 1. Задание (теорема сложения)
# Найти вероятность выпадения 2 или 5 очков при подбрасывании игральной кости, 
# на гранях которой имеются соответственно 1,2,3,4,5 и 6 очков. 
```

Вероятность выпадения 2 очков (событие А):

n (благоприятные исходы) = 1,
N (все исходы) = 6

$P(A) = 1/6$

Вероятность выпадения 5 очков (событие В):

n (благоприятные исходы) = 1,
N (все исходы) = 6

$P(В) = 1/6$

События независимые, применим теорему сложения вероятностей:

$P(A+B) = P(A) + P(B)$,

$P(A+B) = 1/6 + 1/6 = 2/6 = 1/3 = 0.3333%$

Ответ: 0.3333

```python
# 2. Задание (теорема умножения)
# Найти вероятность того, что при двух подбрасываниях той же самой игральной кости сначала выпадет 2, а затем 5.
```

Сначала должна выпасть 2, затем 5.

Вероятность выпадения 2 очков (событие А):

n (благоприятные исходы) = 1,
N (все исходы) = 6

$P(A) = 1/6$

Вероятность выпадения 5 очков (событие В):

n (благоприятные исходы) = 1,
N (все исходы) = 6

$P(В) = 1/6$

События независимые, применим теорему умножения вероятностей:

$P(A*B) = P(A)*P(B)$,

$(A*B) = 1/6 * 1/6 = 1/36 = 0.0277%$

Ответ: 0.0277

```python
# 3. Задание 
# Найти вероятность выпадения 2 и 5 очков при двух подбрасываниях той же самой игральной кости. 
# Обратите внимание на порядок выпадения костей!
```

Событие  выпадения 2 и 5 очков при двух подбрасываниях заключается в выпадении сначала 2, затем 5 очков, либо в выпадении 5, затем 2 очков. 

Обьединив теоремы сложения и умножения вероятностей, получим:

Вероятность наступления события A или B: 

$P(A+B) = P(A*B) + P(B*A)$

$P(A+B) = 1/6 * 1/6 + 1/6 * 1/6 = 2/36 = 1/18 = 0.5555%$

Ответ: 0.5555

```python
# 4. Задание  (Геометрическая вероятность + интервалы)
# На отрезке AB длиной 20 см наугад отметили точку C. Какова вероятность, что она находится на расстоянии 
# не более 9 см от точки А и не более 15 см от точки В?
```

На отрезке AB отложим точку D правее точки А на 9 см, 
точку E левее точки B на 15 см. 

Получим отрезок DE длиной: $AD + BE - AB = 4 см.$

Тогда вероятность равна: $DE/AB = 4/20 = 1/5 = 0.2$

Ответ: 0.2

```python
# 5. Задание. 
# Телефонный номер состоит из 7 цифр. Какова вероятность, что это номер 8882227?
```

Всего способов составить номер:

$9*10^6$ - если номер не может начинаться с 0,

$10^7$ - если номер может начинаться с 0

Тода вероятность равна:

 
$1/(9*10^6)$ - если номер не может начинаться с 0,
 
$1/10^7$ - если номер может начинаться с 0

Ответ: $1/(9*10^6)$; $1/10^7$

```python
# 6. Задание. 
# Набирая номер телефона, абонент забыл 2 последние цифры, и, помня только то, что эти цифры различны 
# и среди них нет нуля, стал набирать их наудачу. Сколько вариантов ему надо перебрать, чтобы наверняка 
# найти нужный номер? Какова вероятность того, что он угадает номер с первого раза?
```

Нужно набрать 2 цифры, каждая из которых может принимать значения от 1 до 9,

Значит количество вариантов которые нужно перебрать = $9^2$ 

Вероятность угадывания номера с первого раза:
    
$P(A) = 1/9^2 = 1/81 = 0.0123$

Ответ: 0.0123
