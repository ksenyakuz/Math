#### 1) Окружность

```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 6, 6


x = []
y1 = []
y2 = []
r = 1
for i in range(-100, 101):
    xx = i/100
    x.append(xx)
    y1.append(np.sqrt(r**2-xx**2))
    y2.append(-np.sqrt(r**2-xx**2))
    
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('X')
plt.ylabel('Y')
```

#### 2) Эллипс

```python
x = []
y1 = []
y2 = []
a = 3
b = 2
for i in range(-300, 301):
    xx = i/100
    x.append(xx)
    y1.append(np.sqrt((1-xx**2/a**2)*b**2))
    y2.append(-np.sqrt((1-xx**2/a**2)*b**2))
    
plt.axis([-4, 4, -4, 4])
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('X')
plt.ylabel('Y')
```

#### 3) Гипербола

```python
y = []
x1 = []
x2 = []
a = 1
b = 1
for i in range(-300, 301):
    yy = i/100
    #if xx <= -a or xx >= a:
    y.append(yy)
    x1.append(np.sqrt((1+yy**2/b**2)/a**2))
    x2.append(-np.sqrt((1+yy**2/b**2)/a**2))
    
plt.axis([-4, 4, -4, 4])
plt.plot(x1, y)
plt.plot(x2, y)
plt.xlabel('X')
plt.ylabel('Y')
```

```python
# 4. Задание (в программе)
# exp(x) + x∙(1 – y) = 1  
# y = x2 – 1
```

```python
x = np.linspace(-2, 3, 201)

plt.figure(figsize = (6, 12))
plt.plot(x, (np.exp(x)+ x - 1)/x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,5) 
plt.grid(True)
plt.axis('scaled')
plt.show()

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (x**2 - 1 - y, np.exp(x) + x * (1 - y) - 1)

x1, y1 =  fsolve(equations, (2, 4))
x2, y2 =  fsolve(equations, (-1, 1))
              
print (x1, y1)
print (x2, y2)
```

```python
x = np.linspace(-2, 3, 201)

plt.figure(figsize = (6, 12))
plt.plot(x, (- 1/x)+2)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,5) 
plt.grid(True)

plt.show()

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (x**2 - 1 - y, (- 1/x)+2 - y)

x1, y1 =  fsolve(equations, (2, 4))
x2, y2 =  fsolve(equations, (0.5, 0.5))
x3, y3 =  fsolve(equations, (-1, 1))
              
print (x1, y1)
print (x2, y2)
print (x3, y3)
```
